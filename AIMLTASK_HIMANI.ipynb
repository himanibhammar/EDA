LINE PLOT

```python
plt.figure(figsize=(55,30))
plt.plot(df['State/UT'],df['Robbery (Sec 392 to 394 IPC)'],label='Robbery')  
plt.title('Robbery Detection', fontdict={'fontname': 'Times New Roman', 'fontsize': 40})
plt.legend()
plt.xlabel('Index')  
plt.ylabel('Count')
plt.show()
```

HISTOGRAM

```python
plt.figure(figsize=(25, 7))
plt.bar(df['State/UT'], df['Robbery (Sec 392 to 394 IPC)'], color='skyblue')
plt.title('Robbery Incidents by State/UT', fontsize=20)
plt.xticks(rotation=45)
plt.ylabel('Count of Robbery Incidents')
plt.xlabel('State/UT')
plt.show()
```

```python

crime_columns = [
    'Murder (Sec.302 IPC)', 
    'Robbery (Sec 392 to 394 IPC)', 
    'Rape (Sec.376 IPC)', 
    'Dacoity (Sec 395 to 398 IPC)'
]

crime_data = df.groupby('State/UT')[crime_columns].sum()
crime_data.plot(kind='bar', stacked=True, figsize=(20, 7))

plt.title('Total Violent Crimes by State/UT', fontsize=20)
plt.ylabel('Count')
plt.xlabel('State/UT')
plt.legend(title='Crime Type')
plt.show()
```

PIE CHART

```python
selected_state = df[df['State/UT'] == 'Odisha']  
crime_distribution = selected_state[crime_columns].sum()

# Pie chart
plt.figure(figsize=(8, 8))
plt.pie(crime_distribution, labels=crime_distribution.index, autopct='%1.1f%%', startangle=140)
plt.title(f'Distribution of Violent Crimes in {selected_state["State/UT"].values[0]}', fontsize=20)
plt.axis('equal')  
plt.show()
```

```python
plt.figure(figsize=(24, 7))
plt.boxplot([df[df['State/UT'] == state]['Robbery (Sec 392 to 394 IPC)'] for state in df['State/UT'].unique()],
            labels=df['State/UT'].unique())
plt.title('Distribution of Robbery Incidents by State/UT', fontsize=20)
plt.xticks(rotation=45)
plt.ylabel('Count of Robbery Incidents')
plt.show()
```
